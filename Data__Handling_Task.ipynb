```python
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
df=pd.read_csv('https://github.com/Taj1920/Data-Analysis-Datasets/raw/refs/heads/main/Datasets-Cleaning/laptopData.csv')
df
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df[df["Company"].isnull()]
# handling missing  values
df.dropna(inplace = True)
df
```

```python
#Remove unnamed column
df=df.drop("Unnamed: 0",axis=1)
df.head()
```

```python
# drop duplicates
df.drop_duplicates(inplace = True)
df.shape
```

```python
# Handling Inches column

df["Inches"]=df["Inches"].replace("?","0")

df["Inches"]=df["Inches"].astype(float)
avg_inch = round(df["Inches"].mean())
avg_inch
df["Inches"]=df["Inches"].replace(0.0,avg_inch)
df["Inches"].unique()
```

```python
# Handling memory column
df["Memory"].unique()
df["Memory"] = df["Memory"].replace("?","256GB SSD")
df["Memory"].unique()
```

```python
df["OpSys"].unique()
```

```python
df["Weight"].unique()
df["Weight"]=df["Weight"].str.strip("kg")
df["Weight"] = df["Weight"].replace("?","0")
df["Weight"] = df["Weight"].astype("float")
avg = round(df["Weight"].mean())
df["Weight"] = df["Weight"].replace(0.0,avg)
df["Weight"].unique()
```

```python
df.reset_index(inplace=True)
df=df.drop("index",axis=1)
df
```

```python
df=pd.read_csv("https://github.com/Taj1920/Data-Analysis-Datasets/raw/refs/heads/main/Datasets-Cleaning/dirty_cafe_sales.csv")
df
```

```python
df["Transaction ID"].unique()
```

```python
df["Item"].unique()
```

```python
df.isnull().sum()
```

```python
df["Item"]=df["Item"].replace(["UNKNOWN", None], "Unknown_Item")
df.isna().sum()
```

```python
df["Quantity"].unique()
```

```python
df["Price Per Unit"]=df["Price Per Unit"].replace(["ERROR","UNKNOWN"],"0")
df["Price Per Unit"] = pd.to_numeric(df["Price Per Unit"], errors='coerce').fillna(0).astype(int)
df["Price Per Unit"].unique()
df.isnull().sum()
```

```python
df["Total Spent"]=df["Total Spent"].replace(["ERROR","UNKNOWN"],"0")
df["Total Spent"] = pd.to_numeric(df["Total Spent"], errors='coerce').fillna(0).astype(int)
df["Total Spent"].unique()
df.isnull().sum()
```

```python
df["Payment Method"]=df["Payment Method"].replace(["ERROR","UNKNOWN"],"0")
df["Payment Method"] = pd.to_numeric(df["Payment Method"], errors='coerce').fillna(0).astype(int)
df["Payment Method"].unique()
df.isnull().sum()
```

```python
df["Location"]=df["Location"].replace(["ERROR","UNKNOWN"],"0")
df["Location"] = pd.to_numeric(df["Location"], errors='coerce').fillna(0).astype(int)
df["Location"].unique()
df.isnull().sum()
```

```python
df["Transaction Date"]=df["Transaction Date"].replace(["ERROR","UNKNOWN"],"0")
df["Transaction Date"] = pd.to_numeric(df["Transaction Date"], errors='coerce').fillna(0).astype(int)
df["Transaction Date"].unique()
df.isnull().sum()
```
