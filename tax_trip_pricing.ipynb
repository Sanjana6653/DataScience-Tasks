```python
Q1 = df["Trip_Distance_km"].quantile(0.25)
Q3 = df["Trip_Distance_km"].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

df = df[(df["Trip_Distance_km"] >= lower_bound) &
        (df["Trip_Distance_km"] <= upper_bound)]
```

```python
from sklearn.preprocessing import LabelEncoder
Le = LabelEncoder()
df["Time_of_Day"]=Le.fit_transform(df["Time_of_Day"])
df["Time_of_Day"]=df["Time_of_Day"].astype(float)
df
```

```python
df["Day_of_Week"]=Le.fit_transform(df["Day_of_Week"])
df["Day_of_Week"]=df["Day_of_Week"].astype(float)
df
```

```python
df['Traffic_Conditions']=Le.fit_transform(df["Traffic_Conditions"])
df["Traffic_Conditions"]=df["Traffic_Conditions"].astype(float)
df
```

```python
df["Weather"]=Le.fit_transform(df["Weather"])
df["Weather"]=df["Weather"].astype(float)
df
```

```python
df.info()
```

```python
X= df.drop('Trip_Price', axis=1)
y=df["Trip_Price"]
```

```python
Xtrain,Xtest,ytrain,ytest = train_test_split(X,y,test_size=0.3,random_state=42)
```

```python
model = LinearRegression()
model.fit(Xtrain,ytrain)
```

```python
print("Intercept : ",model.intercept_)
print("Slope:",model.coef_)
```

MODEL EVALUATION

```python
ypred = model.predict(Xtest)
ypred
```

```python
print("metrics : ")
r2 = r2_score(ytest,ypred)
print("R2 : ",r2)
N = df.shape[0]
P=df.shape[1]
adj_r2 = 1-(((1-r2**2)*(N-1))/(N-P-1))
print("Adjusted R2 : ",adj_r2)
print("MSE : ",mean_squared_error(ytest,ypred))
print("MAE : ",mean_absolute_error(ytest,ypred))
```

```python
print("Training eff :", model.score(Xtrain,ytrain))
print("Testing eff :", model.score(Xtest,ytest))
```
